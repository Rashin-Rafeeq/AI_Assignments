```python
#4.Generate a 1D array of 15 random integers between 1 and 100 and find its min, max, mean, and standard deviation.

import numpy as np
arr = np.random.randint(1, 101, size=15)

minimum = arr.min()
maximum = arr.max()
mean_val = arr.mean()
std_dev = arr.std()

print("Array:", arr)
print("Min:", minimum)
print("Max:", maximum)
print("Mean:", mean_val)
print("Standard Deviation:", std_dev)
```

```python
#5.Given two arrays of sales and costs, calculate profit and profit percentage arrays.

import numpy as np

sales = np.array([1200, 800, 0, 1500, 950])
costs = np.array([700, 500, 100, 1200, 1000])

profit = sales - costs

with np.errstate(divide='ignore', invalid='ignore'):
    profit_pct = np.where(sales != 0, (profit / sales) * 100, np.nan)

profit_rounded = np.round(profit, 2)
profit_pct_rounded = np.round(profit_pct, 2)

print("Sales:         ", sales)
print("Costs:         ", costs)
print("Profit:        ", profit_rounded)
print("Profit %:      ", profit_pct_rounded)
```

```python
#6.Create a 6×6 random matrix and compute row sums, column sums, and total sum.

import numpy as np

matrix = np.random.randint(1, 101, size=(6, 6))

row_sums = matrix.sum(axis=1)
col_sums = matrix.sum(axis=0)
total_sum = matrix.sum()

print("Matrix:\n", matrix)
print("\nRow sums:", row_sums)
print("Column sums:", col_sums)
print("Total sum:", total_sum)
```

```python
#7.Given stock prices, compute daily returns using vectorized NumPy operations.

import numpy as np

prices = np.array([100, 101.5, 99.0, 100.0, 102.0, 101.0], dtype=float)
simple_returns = prices[1:] / prices[:-1] - 1
log_returns = np.log(prices[1:] / prices[:-1])

print("Prices:         ", prices)
print("Simple returns: ", np.round(simple_returns, 6))
print("Log returns:    ", np.round(log_returns, 6))
```

```python
#8.Convert an array of annual interest rates into monthly rates using vectorization.

import numpy as np

annual_rates = np.array([0.03, 0.05, 0.0725, 0.12])
monthly_effective = (1 + annual_rates) ** (1/12) - 1
monthly_nominal = annual_rates / 12

print("Annual rates (%):       ", np.round(annual_rates * 100, 3))
print("Monthly (effective %):  ", np.round(monthly_effective * 100, 4))
```

```python
#9.Normalize a given array of portfolio weights so that the sum becomes 1.

import numpy as np

weights = np.array([0.2, 0.3, 0.15, 0.25, 0.4])

normalized_weights = weights / weights.sum()

print("Original weights:", weights)
print("Normalized weights:", normalized_weights)
print("Sum after normalization:", normalized_weights.sum())
```

```python
#10.Create a Pandas Series from the following list: [1200, 1500, 1800, 2000, 2200]
#a.Find:
#i.mean
#ii.max
#iii.index of the max value

import pandas as pd

data = [1200, 1500, 1800, 2000, 2200]
series = pd.Series(data)

mean_val = series.mean()
max_val = series.max()
max_index = series.idxmax()

print("Series:\n", series)
print("\nMean:", mean_val)
print("Max:", max_val)
print("Index of max value:", max_index)
```

```python
#12.Using this DataFrame:
#df = pd.DataFrame({
 #   "Account": [101,102,103,104],
  #  "Balance": [5000, 15000, 7000, 12000]
#})
#Filter rows where:
#Balance > 8000
#Balance between 6000 and 13000

import pandas as pd

df = pd.DataFrame({
    "Account": [101, 102, 103, 104],
    "Balance": [5000, 15000, 7000, 12000]
})

filter_gt_8000 = df[df["Balance"] > 8000]

filter_between = df[(df["Balance"] >= 6000) & (df["Balance"] <= 13000)]

print("Original DataFrame:\n", df)
print("\nBalance > 8000:\n", filter_gt_8000)
print("\nBalance between 6000 and 13000:\n", filter_between)
```
