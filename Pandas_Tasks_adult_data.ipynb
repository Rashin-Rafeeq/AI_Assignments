Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
df['income_binary'] = np.where(df['income'].str.strip() == '>50K', 1, 0)
display(df.head())
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
bins = [0, 29, 60, np.inf]
labels = ['Youth', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)
display(df.head())
```

Rename the 'educational-num' column to 'education_num'.

```python
df = df.rename(columns={'educational-num': 'education_num'})
display(df.head())
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
df['workclass'] = df['workclass'].replace('?', 'Unknown')
df['occupation'] = df['occupation'].replace('?', 'Unknown')
display(df.head())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = df.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = df.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

Get the count of each marital-status category.

```python
marital_status_counts = df['marital-status'].value_counts()
display(marital_status_counts)
```

Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_per_week_by_education = df.groupby('education')['hours-per-week'].mean()
display(mean_hours_per_week_by_education)
```

Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = df.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

Get the unique values and their counts for the 'native-country' column

```python
native_country_counts = df['native-country'].value_counts()
display(native_country_counts)
```

Filter the dataset to include only individuals with income >50K.

```python
df_high_income = df[df['income'].str.strip() == '>50K']
display(df_high_income.head())
```

Sort the dataset by age in descending order.

```python
df_sorted_by_age = df.sort_values(by='age', ascending=False)
display(df_sorted_by_age.head())
```

Filter the dataset to include only individuals who work more than 40 hours per week.



```python
df_high_hours = df[df['hours-per-week'] > 40]
display(df_high_hours.head())
```

Calculate the percentage of missing values in each column.

```python
missing_values_percentage = (df.isnull().sum() / len(df)) * 100
display(missing_values_percentage)
```

Calculate the median income by education level and gender.

```python
median_income_by_education_gender = df.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

Find the top 3 occupations with the highest average capital-gain.

```python
top_3_occupations_by_capital_gain = df.groupby('occupation')['capital-gain'].mean().nlargest(3)
display(top_3_occupations_by_capital_gain)
```

Filter the dataset to include individuals who are 'Married-civ-spouse' and have an income >50K.

```python
df_married_high_income = df[(df['marital-status'] == 'Married-civ-spouse') & (df['income'].str.strip() == '>50K')]
display(df_married_high_income.head())
```

Create 5 different charts and write your observations from the chart as comment

**Chart 1**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age') #The age distribution is right-skewed, indicating a larger number of younger individuals in the dataset.
plt.ylabel('Frequency') #The most frequent age group appears to be in the 20s to 40s.
plt.show() #There's a significant drop-off in frequency after approximately age 60-70, suggesting fewer older individuals are represented or that the population naturally declines at older ages.
```

**Chart 2**

```python
income_gender_crosstab = pd.crosstab(df['gender'], df['income_binary'], normalize='index') # Create a cross-tabulation of gender and income, normalizing to get proportions

plt.figure(figsize=(8, 6)) # Create a new figure with a specified size
income_gender_crosstab.plot(kind='bar', stacked=True, figsize=(8, 6)) # Plot the cross-tabulation as a stacked bar chart
plt.title('Income Distribution by Gender') # Set the title of the plot
plt.xlabel('Gender') # Set the label for the x-axis
plt.ylabel('Proportion') # Set the label for the y-axis
plt.xticks(rotation=0) # Rotate x-axis labels to prevent overlap
plt.legend(title='Income', labels=['<=50K', '>50K']) # Add a legend for income categories
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show() # Display the plot
```

**Chart 3**

```python
workclass_income_crosstab = pd.crosstab(df['workclass'], df['income_binary'], normalize='index')

plt.figure(figsize=(12, 7))
workclass_income_crosstab.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Income Distribution by Workclass') # 'Self-emp-inc' shows the highest proportion of individuals earning >50K. 'Federal-gov' also has a significant proportion of high earners.
plt.xlabel('Workclass') # 'Never-worked' and 'Without-pay' categories have almost no individuals earning >50K.
plt.ylabel('Proportion') # 'Private' workclass has a lower proportion of high earners compared to some others. 'Unknown' workclass also has a very low proportion of high earners.
plt.xticks(rotation=45, ha='right')
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.tight_layout()
plt.show()
```

**Chart 4**

```python
education_income_crosstab = pd.crosstab(df['education'], df['income_binary'], normalize='index')

plt.figure(figsize=(14, 8))
education_income_crosstab.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Income Distribution by Education Level') # Individuals with higher education levels (e.g., Doctorate, Prof-school, Masters) show a significantly higher proportion of income >50K.
plt.xlabel('Education Level') # Education levels like 'Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', and '12th' have a very low proportion of individuals earning >50K.
plt.ylabel('Proportion') # There is a clear positive correlation between education level and the likelihood of earning >50K.
plt.xticks(rotation=45, ha='right')
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.tight_layout()
plt.show()
```
